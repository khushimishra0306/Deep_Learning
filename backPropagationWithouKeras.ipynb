```python
def initialize_parameters(layers_dims):
    np.random.seed(3)
    parameters={}
    l=len(layers_dims)
    for i in range(1,l):
        parameters['w'+str(i)]=np.ones((layers_dims[i-1],layers_dims[i]))*0.1
        parameters['b'+str(i)]=np.zeros((layers_dims[i],1))
    return parameters
    
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def linear_forward(prev_out,w,b):
    z=np.dot(w.T,prev_out) + b
    A=sigmoid(z)
    return A
```

```python
def forward_propagation(x,parameters):
    A=x
    l=len(parameters)//2

    for i in range(1,l+1):
        A_prev=A
        w1=parameters['w'+str(i)]
        b1=parameters['b'+str(i)]
        #print("A"+str(i-1)+":",A_prev)
        #print("w"+str(i)+":",w1)
        #print("b"+str(i)+":",b1)
        #print("--"*20)


        A=linear_forward(A_prev,w1,b1)
        #print("A"+str(i)+":",A)
        #print("**"*20)

    return A,A_prev
```

```python
def update_parameters(parameters,y,y_hat,x,a1):
    parameters['w2'][0][0]=parameters['w2'][0][0] + (0.001 * (y-y_hat)*a1[0][0])
    parameters['w2'][1][0]=parameters['w2'][1][0] + (0.001 * (y-y_hat)*a1[1][0])
    parameters['b2'][0][0]=parameters['b2'][0][0] + (0.001 * (y-y_hat))


    parameters['w1'][0][0]=parameters['w1'][0][0] + (0.001 * (y-y_hat)*parameters['w2'][0][0])*a1[0][0]*(1-a1[0][0]*x[0][0])
    parameters['w1'][0][1]=parameters['w1'][0][1] + (0.001  * (y-y_hat)*parameters['w2'][0][0])*a1[0][0]*(1-a1[0][0]*x[1][0])
    parameters['b1'][0][0]=parameters['b1'][0][0] + (0.001  * (y-y_hat)*parameters['w2'][0][0])*a1[0][0]*(1-a1[0][0])

    parameters['w1'][1][0]=parameters['w1'][1][0] + (0.001 * (y-y_hat)*parameters['w2'][1][0])*a1[1][0]*(1-a1[1][0]*x[0][0])
    parameters['w1'][1][1]=parameters['w1'][1][1] + (0.001  * (y-y_hat)*parameters['w2'][1][0])*a1[1][0]*(1-a1[1][0]*x[1][0])
    parameters['b1'][1][0]=parameters['b1'][1][0] + (0.001  * (y-y_hat)*parameters['w2'][1][0]) *a1[1][0]*(1-a1[1][0])
```

```python
x=df[['cgpa','profile_score']].values[0].reshape(2,1)
y=df[['lpa']].values[0][0]
```

```python
parameters=initialize_parameters([2,2,1])
y_hat,a1=forward_propagation(x,parameters)
```

```python
y_hat=y_hat[0][0]
```

```python
update_parameters(parameters,y,y_hat,x,a1)
```

```python
loss= -y*np.log(y_hat) - (1-y)*np.log(1-y_hat)
```

```python
print(loss)
```

```python
#epochs implementation
parameters=initialize_parameters([2,2,1])
epochs=50

for i in range(epochs):
    loss=[]

    for j in range(df.shape[0]):
        x=df[['cgpa','profile_score']].values[j].reshape(2,1)
        y=df[['lpa']].values[j][0]
        y_hat,a1=forward_propagation(x,parameters)
        y_hat=y_hat[0][0]
        update_parameters(parameters,y,y_hat,x,a1)
        loss.append(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))
    print('epoch : ',i+1,'loss -',np.array(loss).mean())
parameters
```
