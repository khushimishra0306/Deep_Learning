```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=100,n_informative=1,n_features=2,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,
                        class_sep=15)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,9))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def Perceptron(x,y):
    w1=w2=b=1
    lr=0.1
    for j in range(1000):
        for i in range(x.shape[0]):
            z=w1*x[i][0] + w2*x[i][1]+b
            if z*y[i]<0:
                w1=w1+lr*y[i]* x[i][0]
                w2=w2+lr*y[i]* x[i][1]
                b=b+lr*y[i]
    return w1,w2,b
```

```python
w1,w2,b=Perceptron(x,y)
```
