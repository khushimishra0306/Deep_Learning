```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=100,n_informative=1,n_features=2,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,
                        class_sep=10)
```

```python
y
```

```python
plt.figure(figsize=(12,9))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def step(x):
    if(x>0):
        return 1
    else:
        return 0
```

```python
def Perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    weights=np.ones(x.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(weights,x[j]))
        weights=weights+lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
```

```python
intercept,coef=Perceptron(x,y)
```

```python
coef
```

```python
intercept
```

```python
m= -(coef[0]/coef[1])
b= -(intercept / coef[1])
print(m)
print(b)
```

```python
x_input=np.linspace(-3,3,100)
y_input=m * x_input + b
```

```python
plt.figure(figsize=(12,9))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='summer',s=100)
plt.ylim(-3,2)
plt.show()
```
